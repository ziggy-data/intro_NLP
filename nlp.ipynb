```python
def separa_palavras(lista_tokens):
    lista_palavras= []
    for token in lista_tokens:
        if token.isalpha():
            lista_palavras.append(token)
    return lista_palavras

separa_palavras(palavras_separadas)
```

```python
lista_token = nltk.tokenize.word_tokenize(artigos)
lista_palavras  = separa_palavras(lista_token)
print(f'O Numero de palavra é {len(lista_palavras)}')
```

<b>Agora sabemos quantas palavras existem no nosso artigo mas, não significa que o corretor saberá corrigir isso tudo.</b>

```python
lista_palavras[:5]
```

Podemos perceber que possuimos letras maiusculas em nossos dados e causando uma diferenca entre os outros dados. Vamos normalizar!

```python
def normalizacao(lista_palavras):
    lista_normalizada = []
    for palavra in lista_palavras:
        lista_normalizada.append(palavra.lower())
    return lista_normalizada

lista_normalizada = normalizacao(lista_palavras)
print(lista_normalizada[:5])
```

Agora para remover as repetições da nossa lista usaremos uma built in ``set()``

```python
set(lista_normalizada)
```

```python
print(f'A quantidade de palavras disponíveis para correção são: {len(set(lista_normalizada))}')
```

## Criação do corretor ortográfico

<strong>
Será feito uma divisão em partes a string em "L" e "gica" e precisaremos inserir uma nova letra. ex: "L"+"a"+"gica" e assim por diante. até que será feito com: "L"+"ó"+"gica" e assim teremos nossos candidatos para correção,ou seja, faremos uma generalização de toda essa operação 


Por exemplo nesse caso nosso modelo de NLP teria que inserir letras na nossa palavra digitada equivocadamente.</strong>

```python
lista = 'lgica'
(lista[:1],lista[1:])#lado direito e esquerdo respectivamente
```

```python
palavra_exemplo = 'lgica'

def insere_letras(fatias):
    novas_palavras = []
    letras = 'abcdefghijklmnopqrstuvwxyzàáâãèéêìíîòóôõùúûç'
    for E,D in fatias:
        for letra in letras:
            novas_palavras.append(E + letra + D)
    return novas_palavras
```

```python
def gerador_palavras(palavra):
    fatias = []
    for i in range(len(palavra)+1):
        fatias.append((palavra[:i],palavra[1:]))
    palavras_geradas = insere_letras(fatias)
    return palavras_geradas
```

```python
gerador_palavras(palavra_exemplo)
```

```python
frequencia = nltk.FreqDist(lista_normalizada)
frequencia.most_common()
```

```python
frequencia['lógica']
```

```python
total_palavras = len(lista_normalizada)

def probabilidade(palavra_gerada):
    return frequencia[palavra_gerada]/total_palavras

print(probabilidade('lagica'))
print(probabilidade('logica'))
print(probabilidade('lógica'))
```

```python
def corretor(palavra):
    palavras_geradas = gerador_palavras(palavra)
    palavra_correta = max(palavras_geradas, key=probabilidade)
    return palavra_correta
```

```python
print(corretor(palavra_exemplo))
print(corretor('lógica'))
print(corretor('lagica'))
```
